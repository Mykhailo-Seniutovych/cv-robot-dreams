### Step 0



```python
import cv2
import numpy as np
import time
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = [15, 10]
```

### Step 1

Load an image (any image that contains faces).

```python
img = cv2.imread('../data/challenge-many-people-2.webp')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.subplot(121), plt.imshow(img)
```

### Step 2

Load the dlib face predictor.

```python
casc_path = "../data/haarcascade_frontalface_default.xml"
face_cascade = cv2.CascadeClassifier(casc_path)
```

### Step 3

Run the predictor on your image.

```python
start_time = time.time()
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)
end_time = time.time()
print("faces detected:", len(faces), ", time: ", (end_time - start_time) * 1000, "ms")
```

### Step 4

Draw bounding boxes around the detected faces and plot the image. Use different colour for each face.

```python
result = np.copy(img)
faces_img = []
for (x, y, w, h) in faces: 
    cv2.rectangle(result, (x, y), (x+w, y+h), (0, 255, 0), 3)
    faces_img.append(img[y:y+h, x:x+w, :])
plt.subplot(121), plt.imshow(result)
```
